# Relax Inc. Data Science Challenge Report

#### Objective: 

Identify which factors predict future user adoption.

#### How do we define *user adoption*?

We define an "adopted user" to mean a user who has logged into the product on three separate days in at least one seven-day period.

#### Approach:

Our approach to identify which factors predict user adoption takes 4 main steps:
1. Engineer the target variable
2. Engineer additional predictive features
3. Exploratory Data Analysis
4. Train a tree-based model on the data
5. Check the model's overall performance
6. Check the relative feature importance of each predictive variable, identify the most important ones.

```python
from IPython.display import Image
```

### 1. Engineer the target variable

We needed to use the data in `'takehome_user_engagement.csv'` to engineer the target variable.  This table lists contains a set of time stamps for each time any user logged into the product.  

We calculate whether or not each user has logged into the product on three separate days in at least one seven-day period.  We then join this with the data in `'takehome_users.csv'`.  This gives us our binary response variable for whether or not any given user is an *"adopted user."*

### 2. Engineer additional predictive features


We engineer a number of new features.  The most important of these features is what we called `days_spanned`, which is the number of days between the date a user created their account and the most recent date that the user logged into the account.

For any given user, as `days_spanned` increases, the user will have had more opportunities to log into the account.  So the probability that a user is an adopted user increases with this `days_spanned` variable.

### 3. Exploratory Data Analysis

First, we checked whether or not the target variable is imbalanced.  

We found that only about 11 percent of users are *"adopted user."*

Next, we visualized how the proportion of adopted users varies with some of the features.

Below, we plot how the proportion of adopted users varies with the `days_spanned` variable.

```python
Image('days_spanned_graph.png', width=500)
```

The proportion of adopted users increases with `days_spanned`.

The proportion of adopted users increases with this `days_spanned` variable, as shown in the graph below.

```python
Image('creation_hour_graph.png', width=500)
```

There does not appear to be any strong relationship between the `creation_hour` variable and the proportion of adopted users.

### 4. Train a tree-based model on the data

We train a `XGBClassifier` model with the default parameters. 

Two of the main reasons for using this class of model are:
1. `XGBClassifier` exposes a `feature_importances_` method, which can tell use the relative predictive importance of each variable in the model.
2. `XGBClassifier` does not require much data preprocessing, such as scaling or handling missing values.

### 5. Check the model's overall performance


#### ROC AUC scores:

Train set cross-validation ROC AUC score: 0.995

Test set ROC AUC score: 0.997

#### Accuracy scores:

Train set accuracy score: 0.983

Test set accuracy score: 0.981

#### 5. Check the relative feature importance of each predictive variable, identify the most important ones.

```python
Image('feature_importance.png', width=600)
```

We found that `days_spanned` is the most important predictive feature among the feature set used in our model.

The second most important predictive feature is `last_session_creation_time`, from which `days_spanned` was calculated.

### Conclusions

Although our model shows that `days_spanned` and `last_session_creation_time` are the most important features, it does not make sense to use these features to predict *future* adoption.  

Further analysis is needed to identify features that can predict *future* adoption.
