#### 1. What would you choose as the key metric of success?

#### Profit

I would choose *profit* as the key metric of success.

As a business manager of Ultimate, I should be concerned with whether this experiment of encouraging driver partners to serve both cities ultimately generates losses or generates profits.  

If drivers from one city go to serve in the other city, they will cause Ultimate to incur the toll costs.  If those drivers do not generate enough revenues (for Ultimate) to cover the toll costs, the experiment would be uneconomical.  In such case, it would be a failure from a business perspective.  If those drivers *do* generate enough revenues (for Ultimate) to cover the toll costs, the experiment would be a success.

#### 2. Describe a practical experiment

#### How you will implement the experiment?

First, I would start collecting driver-level  profit data. Also, I would collect data on any relevant explanatory variables that the team recommends.

Once I had collected this data for some time, I would select a random sample of driver partners and offer to reimbursing their toll costs for some trial period.  This sample of driver partners would be the "treatment group."

We could make two comparisons:
1. We could compare the driver-level profits of "treatment group" during the trial period to the driver-level profits of "control group" during the trial period.
2. We could compare the driver-level profits of "treatment group" during the trial period to the driver-level profits of "treatment group" during the pre-trial period.

#### What statistical test(s) you will conduct to verify the significance of the observation?

We could perform a Difference in Means test for each of the aforementioned comparisons.  Specifically, these would be two-sample t-tests.
 


#### How you would interpret the results and provide recommendations to the city operations team along with any caveats?

+ If the driver-level profits of "treatment group" during the trial period were making significant losses, and the Difference in Means test showed there to be a statistically significant difference between the mean driver-level profits of "treatment group" and the driver-level profits of "control group," I would recommend that the city operations teams end the experiment and discontinue the toll reimbursement.

+ If the driver-level profits of "treatment group" during the trial period were *not* making significant losses, and the hypothesis test did *not* show a statistically significant difference in means, I might recommend that the city operations teams continue running experiment for a second trial period.

+ If the driver-level profits of "treatment group" during the trial period were making statistically significant profits, I might recommend making the toll reimbursement available to all driver partners.
