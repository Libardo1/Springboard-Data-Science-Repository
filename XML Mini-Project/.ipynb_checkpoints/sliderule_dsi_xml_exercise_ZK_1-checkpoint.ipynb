## XML example

+ for details about tree traversal and iterators, see https://docs.python.org/2.7/library/xml.etree.elementtree.html

```python
document_tree = ET.parse( './data/mondial_database_less.xml' )
```

```python
# print names of all countries
for child in document_tree.getroot():
    print child.find('name').text
```

```python
# print names of all countries and their cities
for element in document_tree.iterfind('country'):
    print '* ' + element.find('name').text + ':',
    capitals_string = ''
    for subelement in element.getiterator('city'):
        capitals_string += subelement.find('name').text + ', '
    print capitals_string[:-2]
```

Print the last ten thousand characters of the document to see what it looks like.

```python
end = len(etree.tostring(document, encoding='utf-8', pretty_print=True).decode())
start = end - 10000

print(etree.tostring(document, 
                     encoding='utf-8', 
                     pretty_print=True).decode()[start:end])
```

Print the middle ten thousand characters of the document to see what it looks like.

```python
doc_length = len(etree.tostring(document, encoding='utf-8', pretty_print=True).decode())

start = round(doc_length / 2) - round(10000 / 2)
end = round(doc_length / 2) + round(10000 / 2)

print(etree.tostring(document, 
                     encoding='utf-8', 
                     pretty_print=True).decode()[start:end])
```
